```python
from object_detection.utils import label_map_util


def get_classes_and_num_classes(pbtxt_fname):
    from object_detection.utils import label_map_util
    label_map = label_map_util.load_labelmap(pbtxt_fname)
    categories = label_map_util.convert_label_map_to_categories(
        label_map, max_num_classes=90, use_display_name=True)
    category_index = label_map_util.create_category_index(categories)
    class_names = [category['name'] for category in categories]
    num_classes = len(class_names)
    return num_classes, class_names

num_classes, class_names = get_classes_and_num_classes("/content/Capstone-Plant-Disease-2/train/Plant-Disease-dataset-38OY_label_map.pbtxt")
print('Total classes:', num_classes)
print('Class names:', class_names)
```

```python
import re

num_steps = 25000
batch_size = 16


train_tfrecord = "/content/Capstone-Plant-Disease-2/train/Plant-Disease-dataset-38OY.tfrecord"
val_tfrecord = "/content/Capstone-Plant-Disease-2/valid/Plant-Disease-dataset-38OY.tfrecord"
train_label_map ="/content/Capstone-Plant-Disease-2/train/Plant-Disease-dataset-38OY_label_map.pbtxt"
val_label_map ="/content/Capstone-Plant-Disease-2/valid/Plant-Disease-dataset-38OY_label_map.pbtxt"

# Navigate to the target folder
%cd /content/models/ssd_model/
print('writing custom configuration file')

pipeline = '/content/models/ssd_model/ssd_mobilenet_v2_fpnlite_320x320_coco17_tpu-8.config'
fine_tune_ckpt = '/content/models/ssd_model/ssd_mobilenet_v2_fpnlite_320x320_coco17_tpu-8/checkpoint/ckpt-0'

with open(pipeline) as file:
    content = file.read()

with open('pipeline_file.config', 'w') as file:
    content = re.sub('fine_tune_checkpoint: ".*?"', 'fine_tune_checkpoint: "{}"'.format(fine_tune_ckpt), content)
    content = re.sub('(input_path: ".*?)(PATH_TO_BE_CONFIGURED/train)(.*?")', 'input_path: "{}"'.format(train_tfrecord), content)
    content = re.sub('(input_path: ".*?)(PATH_TO_BE_CONFIGURED/val)(.*?")', 'input_path: "{}"'.format(val_tfrecord), content)
    content = re.sub(r'(train_input_reader:\s*{\s*label_map_path:\s*").*?(")', r'\1{}\2'.format(train_label_map), content)
    content = re.sub(r'(eval_input_reader:\s*{\s*label_map_path:\s*").*?(")', r'\1{}\2'.format(val_label_map), content)
    content = re.sub('batch_size: [0-9]+', 'batch_size: {}'.format(batch_size), content)
    content = re.sub('num_steps: [0-9]+', 'num_steps: {}'.format(num_steps), content)
    content = re.sub('num_classes: [0-9]+', 'num_classes: {}'.format(num_classes), content)
    content = re.sub('fine_tune_checkpoint_type: "classification"', 'fine_tune_checkpoint_type: "{}"'.format('detection'), content)

    file.write(content)
```

```python
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from tensorflow.lite.python.interpreter import Interpreter

%matplotlib inline

def list_images(directory, extensions):
    image_files = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(extensions):
                image_files.append(os.path.join(root, file))
    return image_files



def preprocess_image(image, input_shape):


    resized_image = cv2.resize(image, (input_shape[1], input_shape[0]))
    resized_image = resized_image.astype(np.float32)
    resized_image = (resized_image - 127.5) / 127.5  # Normalize image
    preprocessed_image = np.expand_dims(resized_image, axis=0)

    return preprocessed_image


def postprocess_output(output_tensors, image_shape, labels, min_conf, image_path, save_path="/content/save_results", save_outputs=False):

    detected_objects = []
    rectangles = output_tensors[1][0]
    classes = output_tensors[3][0]
    scores = output_tensors[0][0]
    img_H, img_W, _ = image_shape

    for i in range(len(scores)):
        if ((scores[i] > min_conf) and (scores[i] <= 1.0)):
            xmin = int(max(1, (rectangles[i][1] * img_W)))
            xmax = int(min(img_W, (rectangles[i][3] * img_W)))
            ymin = int(max(1, (rectangles[i][0] * img_H)))
            ymax = int(min(img_H, (rectangles[i][2] * img_H)))

            object_name = labels[int(classes[i])]
            if object_name:  # Memeriksa apakah object_name tidak kosong
                print("Object Name:", object_name)
                detected_objects.append([object_name, scores[i], xmin, ymin, xmax, ymax])

    print("Detected Objects:", detected_objects)  # Cetak isi detected_objects untuk memeriksanya

    if save_outputs == True:
        image_file = os.path.basename(image_path)
        base_file, ext = os.path.splitext(image_file)
        text_result = base_file + '.txt'
        text_save_path = os.path.join(save_path, text_result)

        with open(text_save_path,'w') as f:
            for detection in detected_objects:
                f.write('%s %.4f %d %d %d %d\n' % (detection[0], detection[1], detection[2], detection[3], detection[4], detection[5]))

    return detected_objects


def visualize_detection(image, detections):
    for detection in detections:
        xmin, ymin, xmax, ymax = detection[2:]
        cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (10, 255, 0), 2)

        label = f"{detection[0]}: {int(detection[1] * 100)}%"
        labelSize, baseLine = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.5, 2)
        label_ymin = max(ymin, labelSize[1] + 10)
        cv2.rectangle(image, (xmin, label_ymin - labelSize[1] - 10), (xmin + labelSize[0], label_ymin + baseLine - 10),
                      (255, 255, 255), cv2.FILLED)
        cv2.putText(image, label, (xmin, label_ymin - 7), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 2)

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(12, 16))
    plt.imshow(image)
    plt.show()


```

```python

#You can download the test dataset using this link : https://universe.roboflow.com/capstone-saasl/capstone-plant-disease/dataset/2
#The dataset format is XML Pascal VOC

images_path='/content/drive/MyDrive/Dataset/test/'
labels_path='/content/labelmap.txt'
model_path='/content/ssd_mobile_net/detect.tflite'
save_result = "/content/save_results"
threshold = 0.5
test_images = 10
extensions = ('.jpg', '.jpeg', '.png', '.bmp')


with open(labels_path, 'r') as f:
    labels = [line.strip() for line in f.readlines()]

image_files = list_images(images_path, extensions)


os.makedirs(save_result, exist_ok=True)

test_images = random.sample(image_files, test_images)

interpreter = Interpreter(model_path=model_path)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

height = input_details[0]['shape'][1]
width = input_details[0]['shape'][2]

if input_details[0]['dtype'] == np.float32:
    float_input = True

for image_path in test_images:

    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_H, img_W, _ = img.shape

    input_data = preprocess_image(img, (height, width))

    if float_input:
        input_data = (np.float32(input_data) - 127.5) / 127.5

    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()

    output_tensors = [interpreter.get_tensor(output['index']) for output in output_details]


    detections = postprocess_output(output_tensors=output_tensors, image_shape=(img_H, img_W, 3), labels=labels, min_conf=threshold,image_path=image_path, save_outputs=True)

    visualize_detection(img, detections)
```

```python
from tqdm import tqdm


images_path='/content/drive/MyDrive/Dataset/test/'
labels_path='/content/labelmap.txt'
model_path='/content/ssd_mobile_net/detect.tflite'
save_result='/content/mAP_3/input/detection-results'
threshold = 0.5

image_list = list_images(images_path, extensions)
test_images_1 = min(500, len(image_list))

print('Starting inference on %d images...' % test_images_1)

extensions = ('.jpg', '.jpeg', '.png', '.bmp')


with open(labels_path, 'r') as f:
    labels = [line.strip() for line in f.readlines()]

image_files = list_images(images_path, extensions)

os.makedirs(save_result, exist_ok=True)

interpreter = Interpreter(model_path=model_path)
interpreter.allocate_tensors()
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

height = input_details[0]['shape'][1]
width = input_details[0]['shape'][2]

if input_details[0]['dtype'] == np.float32:
    float_input = True


# Loop over test images with tqdm progress bar
for i in tqdm(range(test_images_1), desc="Processing images"):
    # Read image
    image_path = image_files[i]
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_H, img_W, _ = img.shape

    # Preprocess image
    input_data = preprocess_image(img, (height, width))

    if float_input:
        input_data = (np.float32(input_data) - 127.5) / 127.5

    # Perform inference
    interpreter.set_tensor(input_details[0]['index'], input_data)
    interpreter.invoke()

    # Postprocess output
    output_tensors = [interpreter.get_tensor(output['index']) for output in output_details]
    detections = postprocess_output(output_tensors=output_tensors, image_shape=(img_H, img_W, 3), labels=labels, min_conf=threshold,image_path=image_path,save_path = save_result, save_outputs=True)

print('Finished inferencing!')
```
